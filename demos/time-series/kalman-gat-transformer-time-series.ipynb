# Forecasting using Kalman GAT + Transformer


```python
from stellargraph.datasets import METR_LA
from stellargraph.layer.kalman_gat_transformer import KalmanGATTransformer
from tensorflow.keras import Model
```

```python
dataset = METR_LA()
data, adj = dataset.load()
train, test = dataset.train_test_split(data, 0.8)
train, test = dataset.scale_data(train, test)
trainX, trainY, testX, testY = dataset.sequence_data_preparation(12, 1, train, test)

trainX = trainX.transpose((0, 2, 1))  # GAT expects (samples, nodes, seq_len)
testX = testX.transpose((0, 2, 1))
print(trainX.shape)
print(testX.shape)
```

```python
layer = KalmanGATTransformer(seq_len=12, adj=adj, gat_layer_sizes=[4], transformer_layers=1)
x_inp, x_out = layer.in_out_tensors()
model = Model(inputs=x_inp, outputs=x_out)
model.compile(optimizer="adam", loss="mae")
model.fit(trainX, trainY, epochs=1, batch_size=8)
```

```python
pred = model.predict(testX[:1])
print(pred.shape)
```
